```python
# Redefining the function to detect outliers using IQR method
def detect_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column < lower_bound) | (column > upper_bound)]

# Detecting outliers in the specified columns using the IQR method
outliers = {
    'poids_enquete_kg': detect_outliers_iqr(data['poids_enquete_kg']),
    'poids_mg_kg': detect_outliers_iqr(data['poids_mg_kg']),
    'valeur_tot_mg_fr': detect_outliers_iqr(data['valeur_tot_mg_fr']),
    'valeur_moy': detect_outliers_iqr(data['valeur_moy'])
}

# Getting the number of outliers for each column
outliers_count = {col: len(outliers_data) for col, outliers_data in outliers.items()}

outliers_count
```

```python
# Handling missing values

# For categorical columns: replacing NaN with the mode
data['regroupement1'].fillna(data['regroupement1'].mode()[0], inplace=True)
data['regroupement2'].fillna(data['regroupement2'].mode()[0], inplace=True)

# For numerical columns: replacing NaN with the median
numerical_columns_with_na = ['poids_enquete_kg', 'poids_mg_kg', 'valeur_tot_mg_fr', 'valeur_moy']
for col in numerical_columns_with_na:
    data[col].fillna(data[col].median(), inplace=True)

# Checking the remaining missing values in the dataset
remaining_missing_values = data.isnull().sum()

remaining_missing_values
```

```python
# Saving the cleaned data to a new CSV file
cleaned_data_path = "pacificdatavizchallenge_cleaned_data.csv"
data.to_csv(cleaned_data_path, index=False)

cleaned_data_path
```

```python
def main():
    st.title('Analyse des produits')

    # Select year widget
    year = st.sidebar.selectbox('Sélectionnez une année:', data['annee'].unique())

    # Filter data based on selected year
    filtered_data = data[data['annee'] == year]

    # Plotting
    st.subheader(f'Histogramme du poids total des produits pour {year}')
    plt.figure(figsize=(10, 6))
    plt.hist(filtered_data['poids_enquete_kg'].dropna(), bins=30, edgecolor='k')
    plt.xlabel('Poids en kg')
    plt.ylabel('Nombre de produits')
    plt.title(f'Distribution du poids des produits pour {year}')
    st.pyplot(plt.gcf())

if __name__ == '__main__':
    main()
```
